```python
import re
import pickle
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

path = Path(path_to_data)

for file in path.glob("*.pkl"):
    m = re.match(r"(gpt-[\d\.]+(?:-[\w]+)?)_(assist|synth)\.pkl", file.name)
    if not m:
        continue
    model, kind = m.groups()
    # print(model, kind)

    with open(file, "rb") as f:
        data = pickle.load(f)

    s = pd.Series(data)
    print(f"{model} {kind} mean: {s.mean()}" )

    # s.hist(bins=30)
    # plt.title(f"{model} â€” {kind}")
    # plt.xlabel("Value")
    # plt.ylabel("Frequency")
    # plt.show()
```
